```python
import os

# Data Files
CUSTOMER_FILE = 'customer.txt'
WAITER_FILE = 'waiter.txt'
FOOD_ORDER_FILE = 'order_food.txt'
PARCEL_FILE = 'parcel_food.txt'
PAYMENT_FILE = 'payment_bill.txt'

def load_data(file_name):
    if os.path.exists(file_name):
        with open(file_name, 'r') as file:
            return [line.strip().split(' | ') for line in file.readlines()]
    return []

def save_data(file_name, data):
    with open(file_name, 'w') as file:
        for item in data:
            file.write(' | '.join(item) + '\n')

# Customer Management
def add_customer():
    customers = load_data(CUSTOMER_FILE)
    customer_id = str(len(customers) + 1)
    name = input("Enter Customer Name: ").strip()
    age = input("Enter Customer Age: ").strip()
    gender = input("Enter Customer Gender: ").strip()
    contact = input("Enter Customer Contact (Phone or Email): ").strip()
    welcome_juice = input("Enter Customer Welcome Juice Name: ").strip()

    customers.append([customer_id, name, age, gender, contact, welcome_juice])
    save_data(CUSTOMER_FILE, customers)
    print(f"Customer {name} has been added.")

def view_customers():
    customers = load_data(CUSTOMER_FILE)
    if not customers:
        print("No Customers Found.")
        return
    print("List of Customers:")
    for customer in customers:
        if len(customer) >= 6:
            print(f"Customer ID: {customer[0]}, Name: {customer[1]}, Age: {customer[2]}, Gender: {customer[3]}, Contact: {customer[4]}, Welcome Juice: {customer[5]}")
        else:
            print(f"Invalid Customer Record: {customer}")

# Waiter Management
def add_waiter():
    waiters = load_data(WAITER_FILE)
    waiter_id = str(len(waiters) + 1)
    name = input("Enter Waiter Name: ").strip()
    work_shift = input("Enter Waiter Work Shift: ").strip()
    gender = input("Enter Waiter Gender: ").strip()
    contact = input("Enter Waiter Contact (Phone or Email): ").strip()

    waiters.append([waiter_id, name, work_shift, gender, contact])
    save_data(WAITER_FILE, waiters)
    print(f"Waiter {name} has been added.")

def view_waiters():
    waiters = load_data(WAITER_FILE)
    if not waiters:
        print("No Waiters Found.")
        return
    print("List of Waiters:")
    for waiter in waiters:
        if len(waiter) >= 5:
            print(f"Waiter ID: {waiter[0]}, Name: {waiter[1]}, Work Shift: {waiter[2]}, Gender: {waiter[3]}, Contact: {waiter[4]}")
        else:
            print(f"Invalid Waiter Record: {waiter}")

# Food Order Management
def order_food():
    orders = load_data(FOOD_ORDER_FILE)
    order_id = str(len(orders) + 1)
    customer_id = input("Enter Customer ID: ").strip()
    customer_name = input("Enter Customer Name: ").strip()
    waiter_id = input("Enter Waiter ID: ").strip()
    waiter_name = input("Enter Waiter Name: ").strip()
    foods_id = input("Enter Foods ID: ").strip()
    foods_name = input("Enter Foods Name: ").strip()
    date = input("Enter Food Order Date (DD-MM-YYYY): ").strip()
    time = input("Enter Food Order Time (HH:MM): ").strip()

    orders.append([order_id, customer_id, customer_name, waiter_id, waiter_name, foods_id, foods_name, date, time])
    save_data(FOOD_ORDER_FILE, orders)
    print("Food order has been successful.")

def view_food_orders():
    orders = load_data(FOOD_ORDER_FILE)
    if not orders:
        print("No Food Orders Found.")
        return
    print("List of Food Orders:")
    for order in orders:
        if len(order) >= 9:
            print(f"Order ID: {order[0]}, Customer ID: {order[1]}, Customer Name: {order[2]}, Waiter ID: {order[3]}, Waiter Name: {order[4]}, Foods ID: {order[5]}, Foods Name: {order[6]}, Date: {order[7]}, Time: {order[8]}")
        else:
            print(f"Invalid Food Order Record: {order}")

# Parcel Food Management
def parcel_food():
    parcels = load_data(PARCEL_FILE)
    parcel_id = str(len(parcels) + 1)
    customer_id = input("Enter Customer ID: ").strip()
    customer_name = input("Enter Customer Name: ").strip()
    waiter_id = input("Enter Waiter ID: ").strip()
    waiter_name = input("Enter Waiter Name: ").strip()
    foods_id = input("Enter Parcel Foods ID: ").strip()
    foods_name = input("Enter Parcel Foods Name: ").strip()
    date = input("Enter Food Parcel Date (DD-MM-YYYY): ").strip()
    time = input("Enter Food Parcel Time (HH:MM): ").strip()

    parcels.append([parcel_id, customer_id, customer_name, waiter_id, waiter_name, foods_id, foods_name, date, time])
    save_data(PARCEL_FILE, parcels)
    print("Food parcel has been successful.")

def view_food_parcels():
    parcels = load_data(PARCEL_FILE)
    if not parcels:
        print("No Food Parcels Found.")
        return
    print("List of Food Parcels:")
    for parcel in parcels:
        if len(parcel) >= 9:
            print(f"Parcel ID: {parcel[0]}, Customer ID: {parcel[1]}, Customer Name: {parcel[2]}, Waiter ID: {parcel[3]}, Waiter Name: {parcel[4]}, Foods ID: {parcel[5]}, Foods Name: {parcel[6]}, Date: {parcel[7]}, Time: {parcel[8]}")
        else:
            print(f"Invalid Food Parcel Record: {parcel}")

# Payment Management
def bill_payment():
    payments = load_data(PAYMENT_FILE)
    payment_id = str(len(payments) + 1)
    order_id = input("Enter Order ID: ").strip()
    customer_id = input("Enter Customer ID: ").strip()
    total_amount = input("Enter Total Amount: ").strip()
    payment_method = input("Enter Payment Method (Cash/Card): ").strip()
    payment_date = input("Enter Payment Date (DD-MM-YYYY): ").strip()

    payments.append([payment_id, order_id, customer_id, total_amount, payment_method, payment_date])
    save_data(PAYMENT_FILE, payments)
    print("Payment has been processed successfully.")

def view_payments():
    payments = load_data(PAYMENT_FILE)
    if not payments:
        print("No Payments Found.")
        return
    print("List of Payments:")
    for payment in payments:
        if len(payment) >= 6:
            print(f"Payment ID: {payment[0]}, Order ID: {payment[1]}, Customer ID: {payment[2]}, Total Amount: {payment[3]} Taka, Payment Method: {payment[4]}, Payment Date: {payment[5]}")
        else:
            print(f"Invalid Payment Record: {payment}")

def foods_menu():
    print("\nFoods List:")
    foods = ["1. Rice ~80.00 $(1)", "2. Beef ~190.00 $(1)", "3. Mutton ~220.00 $(1)", "4. Kacci Biriyani ~340.00 $(1)","5. Fuska ~50.00 $(1)","6. Jalmori ~20.00 $(1)"]
    for food in foods:
        print(food)

# Main Menu
def main():
    while True:
        print("\n...Nur Muhammad Five Star Hotel....")
        print("\n...............Restaurant Management System .............")
        print("1. Add Customer")
        print("2. View All Customers")
        print("3. Add Waiter")
        print("4. View All Waiters")
        print("5. Food Order")
        print("6. View All Food Orders")
        print("7. Food Parcel")
        print("8. View All Food Parcels")
        print("9. See Food Menu")
        print("10. Bill Payment")
        print("11. View All Payments")
        print("12. Exit")

        choice = input("Enter Your Choice: ").strip()

        if choice == '1':
            add_customer()
        elif choice == '2':
            view_customers()
        elif choice == '3':
            add_waiter()
        elif choice == '4':
            view_waiters()
        elif choice == '5':
            order_food()
        elif choice == '6':
            view_food_orders()
        elif choice == '7':
            parcel_food()
        elif choice == '8':
            view_food_parcels()
        elif choice == '9':
            foods_menu()
        elif choice == '10':
            bill_payment()
        elif choice == '11':
            view_payments()
        elif choice == '12':
            print("Good Bye!!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 12.")

if __name__ == "__main__":
    main()
```
